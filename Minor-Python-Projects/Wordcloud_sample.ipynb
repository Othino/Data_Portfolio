```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
#the sample text to be coverted to image
sample_text = '''Learning how to scrape data using Power Automate and Power BI is a game changer.Good thing is that it only takes minutes. 
You are required to have a developer account in Microsoft 365 to use Power Automate and Power BI'''
```

```python
wcd= WordCloud(width = 200, height = 200,background_color="white").generate(sample_text)

# to remove the axis and show the data as an image
plt.axis("off")
plt.imshow(wcd, interpolation = "bilinear")
```

```python
wcd = WordCloud(width = 1000, height = 1000, repeat=True).generate(sample_text)

# to remove the axis and display the data as image
plt.axis("off")
plt.imshow(wcd, interpolation = "bilinear")
```
