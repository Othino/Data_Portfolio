#Plotting

```python
sns.countplot(x ='Marital_Status', data = marketing)
```

The data implies that a higher number of customers are a couple as compared to those who are alone.

```python
sns.countplot(x ='Education', data = marketing)
```

This data shows that majority of the customers have attained postgraduate education level

```python
sns.kdeplot(x = 'Age' , data = marketing , color = 'blue')
```

The peak of the above graph is in between 40 and 60 so we can conclude that most customers were between the ages of 40 and 60.


```python
sns.countplot(x= 'Children', data = marketing)
```

This data shows that the majority of customers have 1 child while the least have 3 children.

```python
sns.kdeplot(x = 'Income' , data = marketing , color = 'blue')
```

The plot shows majority of the customers earn between 0-100000, however, a minority earn between 100000 to 200000 as shown from the slight spike between that range.

```python
sns.displot(x = 'Expenditure',kde=True,bins = 5 , data =marketing)
```

This plot shows that most of the customers spend approximately $500 when purchasing items from the business.


```python
sns.countplot(x= 'Accepted campaigns', data = marketing)
```

The plot shows that majority of customers did not accept the campaigns run by the business.

```python
sns.countplot (x = 'NumDealsPurchases', data=marketing)
```

This plot shows that most customers made only 1 purchase with given discounts. This raises the question on why while the deals are running, customers were adamant to use the discount offered by the business. 



```python
purchase_chart = marketing[['Web', 'Catalog', 'Store']].sum().sort_values(ascending=True).plot(kind='barh', title = 'Purchases')
```

This data shows that majority of the customers purchase from the physical store while the least people purchase from the catalog. The low purchases made from the catalog might be because the products included do not match customers' specific needs.

```python
Expense_chart = marketing[['Wine','Fruits','Meat_Products','Fish_Products','GoldProds','Sweet_Products']].sum().sort_values(ascending=False).plot(kind='bar', title ='Expenses')
```

The data shows that most of the customers spend money on wine while the least spend money on fruits

```python
Customer_recency = marketing.head(10).plot.line(y = 'Recency')
```

This data shows the the most recent times a customer shopped from the business.

```python
sns.countplot(x ='Customer_year', data = marketing)
```

The plot shows that most of the customers have been buying from the business for 9 years.

```python
Education_expense = marketing[['Education','Expenditure']].groupby('Education').sum().sort_values(by='Expenditure', ascending=False).plot(kind='bar', figsize=(5,5), legend=None, color='blue')
```

The plot shows that the highest spenders are consumers with postgraduate education level

```python
df5 = marketing[['Marital_Status','Expenditure']].groupby('Marital_Status').sum().sort_values(by='Expenditure', ascending=False).plot(kind='bar', figsize=(5,5), legend=None, color='blue')
```

The plot shows that highest spenders from the business are couples rather than single people.

```python
df6 = marketing[['Children','Expenditure']].groupby('Children').sum().sort_values(by='Expenditure', ascending=False).plot(kind='bar', figsize=(5,5), legend=None, color='blue')
```

The plot shows that people without children have the highest expenditure in the business.

```python
df7 =marketing [['Children','Accepted campaigns']].groupby('Children').sum().sort_values(by='Accepted campaigns', ascending=False).plot(kind='bar', figsize=(5,5), legend=None, color='blue')
```

Customers without kids do not accept the marketing campaigns offered by the business

```python
df8 = marketing[['Marital_Status','Accepted campaigns']].groupby('Marital_Status').sum().sort_values(by='Accepted campaigns', ascending=False).plot(kind='bar', figsize=(5,5), legend=None, color='blue')
```

The data shows that customers who are a couple accept campaign offers from the business

```python
df9 = marketing[['Education','Accepted campaigns']].groupby('Education').sum().sort_values(by='Accepted campaigns', ascending=False).plot(kind='bar', figsize=(5,5), legend=None, color='blue')
```

The data shows that customers with postgraduate education level accept the campaigns as compared to those from undergraduate level

**CONCLUSION**

Based on the above analysis, some of the insights derived from are as follows:

•	Majority of the customers are postgraduate students, are couple and have one child.

•	Customers prefer buying from the physical store rather than through the website or catalogs sent via email.

•	Majority of the customers purchase wine products from the business.

•	The business has more older customers and no new customers

•	Higher levels of expenditure come from consumers who have postgraduate education level, are a couple and have no children.

•	Additionally, most customers that accepted the campaigns run by the business had no children, were a couple and had postgraduate degree.

Thus, for the business, this means that:

•	When designing marketing campaigns, the business should target customers with post graduate degree, coupled up and have no children

•	The marketing campaigns should be tailored towards promoting wine since it is the product spent on by majority of consumers.

•	Marketing campaigns to attract new customers should be implemented to drive sales in the business.
